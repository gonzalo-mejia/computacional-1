```python
from numpy import array,cos,sin
from numpy import loadtxt
import numpy as np
import scipy as sc
import matplotlib.pyplot as plot
from pylab import plot, show,plt
```

```python
#us=1.495*10**11
m=1
gm=(4*pow(3.1416,2))
dt=0.02
n=200
r=1
v=2*(3.1416)
the=0
V=[]
R=[]
t=[]
ET=[]
x=[]
y=[]
THE=[]
for i in range (0,n):
    r=r+(dt*v)
    v=v+(gm/(r**2))*dt
    E=-((gm*m)/(2*r))
    the=the+(v/r)*dt
    V.append(v)
    R.append(r)
    THE.append(the)
    x.append(r*cos(the))
    y.append(r*sin(the))
    t.append(i*dt)
    ET.append(E)
    #print(V)
plt.polar(THE,R,'o',color='aqua')
plt.show()
#plot(t,ET,color='black')
```
